```python
import requests

import time

data_points = []

for _ in range(100):
    response = requests.get("http://api.open-notify.org/iss-now.json")
    data = response.json()
    
    latitude = data["iss_position"]["latitude"]
    longitude = data["iss_position"]["longitude"]
    timestamp = data["timestamp"]
    
    data_points.append({
        "latitude": latitude,
        "longitude": longitude,
        "timestamp": timestamp
    })
    
    time.sleep(1)

for data_point in data_points:
    print("Latitude:", data_point["latitude"])
    print("Longitude:", data_point["longitude"])
    print("Timestamp:", data_point["timestamp"])
    print("-" * 20)
```
